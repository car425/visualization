```python
# Import necessary libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Generate "data" for a figure.  (Random assortment of data points).

np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize = (5, 3)) # Make a figure (container or canvas) with our axes (object, graph itself).  
# This line says to make a figure and then put an axes in that figure.  Next line specifies that axes is a scatter plot.
ax.scatter(x, y)
```

```python
# Presenting the same data in the above cell as a bar plot instead of a scatter plot.

fix, ax = plt.subplots(figsize = (5, 3))
ax.bar(x, y)
```

```python
# Repeating the step of generating "data" for a figure.  (Random assortment of data points).

np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

# Plotting that data as a line graph

fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y)
```

```python
# Plotting the same data as a histogram.  Note that only the y axis needs to be specified.

fig, ax = plt.subplots(figsize = (5, 3))
ax.hist(y)
```

```python
# Learning how to add labels and a title to the figure

fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
# Performing more specific modifications to the line graph.

# Make a font dictionary for our labels

font1 = {'family': 'sans-serif', 'color': 'blue', 'size':20}
font2 = {'family': 'monospace', 'color': 'green', 'size':14}

fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y)

# Call on these font dictionaries to modify our figure

ax.set_title('Total growth over time', fontdict= font1)
ax.set_ylabel('Total growth', fontdict= font2)
ax.set_xlabel('Years since start',  fontdict= font2)
```

```python
# Make a font dictionary for our labels

font1 = {'family': 'sans-serif', 'color': 'blue', 'size':20}
font2 = {'family': 'monospace', 'color': 'green', 'size':14}

fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y)

# Call on these font dictionaries to modify our figure

ax.set_title('Total growth over time', fontdict= font1, loc = 'left') # Locking title into left alignment.  Default is center alignment.
ax.set_ylabel('Total growth', fontdict= font2)
ax.set_xlabel('Years since start',  fontdict= font2)
```

```python
# Customizing plot appearance

fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y, marker = '*', color = 'indigo')

# Have specified a star-shaped marker.
```

```python
# Customizing plot appearance

fig, ax = plt.subplots(figsize = (5, 3))
ax.scatter(x, y, marker = '*', color = 'yellow')
ax.set_facecolor('black')

# I made the night sky!
```

```python
# Further customizing plot appearance by specifying details about the connecting lines.

fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y, 
        marker = '*', 
        color = 'indigo',
        linestyle = '--', 
        linewidth = 2)
```

```python
# More customization.  Using hexcodes to optimize our colors

fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y, 
        marker = '*', 
        markersize = 12,
        color = 'indigo',
        linestyle = '--', 
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')
```

```python
# Adding grid lines to axes object. Default is to modify both 'x' and 'y' gridlines.

fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y, 
        marker = '*', 
        markersize = 12,
        color = 'indigo',
        linestyle = '--', 
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')

ax.grid()
```

```python
# Adding grid lines to axes object.  ('y' for horizontal gridlines)

fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y, 
        marker = '*', 
        markersize = 12,
        color = 'indigo',
        linestyle = '--', 
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')

ax.grid(axis = 'y')
```

```python
# Adding grid lines to axes object.  ('x' for horizontal gridlines)

fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y, 
        marker = '*', 
        markersize = 12,
        color = 'indigo',
        linestyle = '--', 
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')

ax.grid(axis = 'x')
```

```python
# Testing additional customization options

fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(x, y, 
        marker = '*', 
        markersize = 12,
        color = 'indigo',
        linestyle = '--', 
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')

ax.grid(axis = 'y', color = 'blue', linestyle = '-', linewidth = 0.5)
```

```python
# Create a new heatmap from Python Graph Gallery
# https://python-graph-gallery.com/90-heatmaps-with-various-input-format/

# library
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(613)
 
# Create a dataset
df = pd.DataFrame(
    np.random.random((5,5)),
    columns=["a","b","c","d","e"]
)

# Default heatmap: just a visualization of this square matrix
sns.heatmap(df)
plt.show()
```

```python
# Calculate correlation between each pair of variable
corr_matrix=df.corr()
 
# plot it
sns.heatmap(corr_matrix, cmap='PuOr')
plt.show()
```

```python
# Playing with scatterplots from https://python-graph-gallery.com/scatter-plot/
# library & dataset
import seaborn as sns
df = sns.load_dataset('iris')

# use the function scatterplot() to make a scatterplot
sns.scatterplot(x=df["sepal_length"], y=df["sepal_width"],
                color = 'green',
                edgecolor = 'blue',
                alpha = 0.5, 
                s = 20)

# Transparency of all points is the same, but some darker "points" are due to multiple points overlapping in the plot.
```
