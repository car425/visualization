```python
# Import libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy
import PIL
import requests

# Generate "data"

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
# Plotting both y variables on same x axis

fig, ax = plt.subplots(figsize = (5, 3)) 
ax.plot(x, y1, y2)
```

```python
# Plotting both y variables on same x axis

fig, ax = plt.subplots(figsize = (5, 3)) 

# Plot two y axes differently so that they can later be modified separately.
ax.plot(x, y1)
ax.plot(x, y2)

# Optional, sometimes necessary to make graph appear.
fig.show()
```

```python
# Plotting both y variables on same x axis

fig, ax = plt.subplots(figsize = (5, 3)) 

# Plot two y axes differently so that they can later be modified separately.
ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = 'Person 2')
ax.legend(loc = 'lower right')

# Optional, sometimes necessary to make graph appear.
fig.show()
```

```python
# Further customization
# # Plotting both y variables on same x axis

fig, ax = plt.subplots(figsize = (5, 3)) 

# Plot two y axes differently so that they can later be modified separately.
ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = 'Person 2')
ax.legend(loc = 'lower right',
          frameon = True,
          fontsize = 12,
          ncol = 2,
          shadow = True)

# ncol is number of columns in legend

# Optional, sometimes necessary to make graph appear.
fig.show()
```

```python
# Further customization
# # Plotting both y variables on same x axis

fig, ax = plt.subplots(figsize = (5, 3)) 

# Plot two y axes differently so that they can later be modified separately.
ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = 'Person 2')
ax.legend(loc = 'upper left',
          bbox_to_anchor = (1, 1))

# bbox_to_anchor will tell bounding box where to anchor.  Code means to anchor upper left corner of legend.  
# (1, 1) means anchored as far to the right and as far to the top as possible on axes.  It is a coordinate.

# Optional, sometimes necessary to make graph appear.
fig.show()
```

```python
# Working with text and annotations

fig, ax = plt.subplots(figsize = (5, 3)) 

# Plot two y axes differently so that they can later be modified separately.
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc = 'lower right')

# How to annotate a graph with added text
# Coordinates (10, 95) are specified.  They are in the units of our plot.

ax.text(10, 95, 'This value is important!')
fig.show()
```

```python
# Working with text and annotations

fig, ax = plt.subplots(figsize = (5, 3)) 

# Plot two y axes differently so that they can later be modified separately.
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc = 'lower right')

# How to annotate a graph with added text
# Coordinates (10, 95) are specified.  They are in the units of our plot.

# Testing additional modifications to text

ax.text(10, 95, 'This value is important!',
        ha = 'center',   # alignment of text.  ha = horizontal alignment.  Centered around coordinates 10, 95.  Could also use "left" or "right"
        color = 'red',  # modify font color
        size = 20)      # modify font size
fig.show()
```

```python
# Working with text and annotations
# Using transforms to position text annotations

# Make a simple, empty set of axes

fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10]) # goes from 0 to 10 horizontally, and 0 to 10 vertically.

# Each transform tells us how to specify the coordinates

ax.text(1, 5, '.Data: (1, 5)', transform = ax.transData)  # Coordinates are in terms of data on the chart.
ax.text(0.5, 0.1, '.Axes: (0.5, 0.1)', transform = ax.transAxes) # Specify location on axes object.  Relative position on axes image (50% horizontally and 10% vertically)
ax.text(0.2, 0.2, '.Figure: (0.2, 0.2)', transform = fig.transFigure) # Specify location on figure (canvas).  Relative position is 20%, 20%.

#fig.show()
```

```python
# Can use arrows to annotate

fig, ax = plt.subplots(figsize = (5, 3)) 

# Plot two y axes differently so that they can later be modified separately.
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc = 'lower right')

ax.annotate('This is important!', xy = (10, 95,), xytext = (20, 94), arrowprops= dict(facecolor = 'black'))  

#xy = (10, 95) specifies the coordinate we're pointing at
#xytext specifies coordinates of text.
# arrowprops modifies the arrow itself.
```

```python
# Can use arrows to annotate
# Change the arrow!!!

fig, ax = plt.subplots(figsize = (5, 3)) 

# Plot two y axes differently so that they can later be modified separately.
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc = 'lower right')

ax.annotate('This is important!', xy = (10, 95,), xytext = (20, 94), arrowprops= dict(facecolor = 'hotpink', arrowstyle = 'wedge', edgecolor = 'blue'))  

#xy = (10, 95) specifies the coordinate we're pointing at
#xytext specifies coordinates of text.
# arrowprops modifies the arrow itself.
```

```python
# Removing tick marks and labels

fig, ax = plt.subplots(figsize = (5, 3)) 

# Plot two y axes differently so that they can later be modified separately.
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc = 'lower right')

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# Subplots and multiple axes

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy
import PIL
import requests

# A figure is a container that holds a set of axes
# axes is our actual plot or graph
# Every visual element of our plots - color, legends, axis titles and scales, text - is called and artist and belongs to an axes (not to a figure)
```

```python
# Generate "data"

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji']) # Categorical data
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# Multiple axes in a single figure

fig, (ax1, ax2) = plt.subplots(ncols = 2, 
                                nrows = 1, 
                                figsize = (7, 3))
ax1.scatter(x1, y1)
ax2.bar(x2, y2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, 
                                nrows = 1, 
                                figsize = (7, 3))
ax1.scatter(x1, y1, marker = '*', facecolor = 'green',  s = 20)
ax2.bar(x2, y2, color = 'turquoise')
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                   ['ax2', 'ax3']], 
                                   figsize = (7, 4))

# What each axes object looks like

someaxes["ax1"].scatter(x1, y1)
someaxes["ax2"].scatter(x2, y2)
someaxes["ax3"].plot(x1, y1)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                   ['ax2', 'ax4']], 
                                   figsize = (7, 4))

# What each axes object looks like

someaxes["ax1"].scatter(x1, y1)
someaxes["ax2"].scatter(x2, y2)
someaxes["ax3"].plot(x1, y1)
someaxes["ax4"].plot(x1, y1, color = 'green')

# 'someaxes' is a name for a group of axes.  It allows us to work with a group of axes objects.
```

```python
fig, myaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                   ['ax2', 'ax4']], 
                                   figsize = (7, 4))

# What each axes object looks like

myaxes["ax1"].scatter(x1, y1)
myaxes["ax2"].scatter(x2, y2)
myaxes["ax3"].plot(x1, y1)
myaxes["ax4"].plot(x1, y1, color = 'green')

# 'someaxes' is a name for a group of axes.  It allows us to work with a group of axes objects.
```

```python
# Add x axis labels to each plot.

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                   ['ax2', 'ax3']], 
                                   figsize = (7, 4))

# What each axes object looks like

someaxes["ax1"].scatter(x1, y1)
someaxes["ax2"].bar(x2, y2)
someaxes["ax3"].plot(x1, y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes['ax2'].set_xlabel('Another Label', fontsize = 18)
someaxes['ax3'].set_xlabel('Label 2 : 2 Fast 2 Furious', fontsize = 18)
```

```python
# Fixing above issues using layouts

# Tight layout 

# Constrained layout

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                   ['ax2', 'ax3']], 
                                   figsize = (7, 4), layout = 'constrained')

# What each axes object looks like

someaxes["ax1"].scatter(x1, y1)
someaxes["ax2"].bar(x2, y2)
someaxes["ax3"].plot(x1, y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes['ax2'].set_xlabel('Another Label', fontsize = 18)
someaxes['ax3'].set_xlabel('Label 2 : 2 Fast 2 Furious', fontsize = 18)
```

```python
# Fixing above issues using layouts

# Tight layout 

# Constrained layout

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                   ['ax2', 'ax3']], 
                                   figsize = (7, 4), layout = 'tight')

# What each axes object looks like

someaxes["ax1"].scatter(x1, y1)
someaxes["ax2"].bar(x2, y2)
someaxes["ax3"].plot(x1, y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes['ax2'].set_xlabel('Another Label', fontsize = 18)
someaxes['ax3'].set_xlabel('Label 2 : 2 Fast 2 Furious', fontsize = 18)

#plt.tight_layout()
```

```python
# Multiple viz on one axes object

# Generate "data"

x = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji']) # Categorical data
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

fig, ax = plt.subplots(figsize = (7, 3))

# Layer two visuals onto a single axes object

ax.bar(x, y1, color = 'indigo')
ax.plot(x, y2, color = 'red')
```

```python
# Add error bars

# Calculate std deviation
y2_sd = np.std(y2)

# Write function

fig, ax = plt.subplots(figsize = (7, 3))
ax.plot(x, y2, color = 'red')

ax.errorbar(x, y2,
            yerr = y2_sd, 
            fmt = 'none')  # allows us to not have to use default format for error bars.  See below with this argument excluded.
```

```python
fig, ax = plt.subplots(figsize = (7, 3))
ax.plot(x, y2, color = 'red')

ax.errorbar(x, y2,
            yerr = y2_sd)  
```

```python
fig, ax = plt.subplots(figsize = (7, 3))
ax.plot(x, y2, color = 'red')

ax.errorbar(x, y2,
            yerr = y2_sd, 
            fmt = 'none',
            ecolor = 'indigo', #ecolor means error color
            elinewidth = 4, 
            capsize = 6, 
            capthick = 4)  
```

```python
fig, ax = plt.subplots(figsize = (7, 3))
ax.plot(x, y2, color = 'red')

ax.errorbar(x, y2,
            yerr = y2_sd, 
            
            ecolor = 'indigo', #ecolor means error color
            elinewidth = 4, 
            capsize = 6, 
            capthick = 4)  
```

```python
# Only certain points have error bars.  Specify interval.

fig, ax = plt.subplots(figsize = (7, 3))
ax.plot(x, y2, color = 'red')

ax.errorbar(x, y2,
            yerr = y2_sd, 
            fmt = 'none',
            ecolor = 'indigo', #ecolor means error color
            elinewidth = 4, 
            capsize = 6, 
            capthick = 4,
            errorevery= 2)  
```
