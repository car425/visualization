```python
# Loading a dataset from seaborn

tips = sns.load_dataset('tips')
print(tips)
```

```python
# Making a plot with seaborn

sns.lineplot(data = tips, 
             x = 'total_bill', 
             y = 'tip')
```

```python
# Changing plot features with seaborn

sns.set_style('darkgrid')

sns.lineplot(data = tips, 
             x = 'total_bill', 
             y = 'tip')
```

```python
# Add title and axis labels.  Make lineplot into object called tipgraph and then specify what we want to add.

tipgraph = sns.lineplot(data = tips, 
             x = 'total_bill', 
             y = 'tip')
tipgraph.set(title = 'Tips vs Total Bill', 
             xlabel = 'Total Bill ($)', 
             ylabel = 'Tip Amount ($)')
```

```python
# Further customizing plot

fig = plt.subplots(figsize = (10, 3))

tipgraph = sns.lineplot(data = tips, 
             x = 'total_bill', 
             y = 'tip',
             color = 'hotpink', 
             linestyle = '--',
             linewidth = 3,
             marker = 'o', 
             markerfacecolor = 'indigo')
tipgraph.set(title = 'Tips vs Total Bill', 
             xlabel = 'Total Bill ($)', 
             ylabel = 'Tip Amount ($)')

plt.show()
```

```python
# Altering plot so that different times of day and different days have different colors and symbols.

tipgraph = sns.scatterplot(data = tips, 
             x = 'total_bill', 
             y = 'tip',
             style = 'time',
             hue = 'day',
             palette = ['purple', 'hotpink', 'deepskyblue', 'yellowgreen'])

tipgraph.set(title = 'Tips vs Total Bill', 
             xlabel = 'Total Bill ($)', 
             ylabel = 'Tip Amount ($)')

plt.show()
```

```python
# Pairplot with Seaborn for exploratory data analysis

sns.pairplot(data = tips, hue = 'day')
```

```python
# Compare levels within variables with relplot

daysplot = sns.relplot(
    data = tips,
    x = 'total_bill', 
    y = 'tip', 
    hue = 'sex',  # Differentiate hue by sex
    col = 'day',  # Differentiate column by day.  Subplots in columns arranged by day.
    kind = 'scatter',  # Use a scatter plot
    col_wrap = 2) # Set number of columns

daysplot

```

```python
import plotly.graph_objects as go  # "go is graph objects"
import nbformat
```

```python
# Categorical array of names and values

x1 = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y1 = np.array ([110, 180, 240, 99, 220])
```

```python
# Make a plot using plotly.  Figure will be interactive.

graph = go.Figure() # Make our figure
graph.add_trace(go.Bar(x=x1, y = y1)) # Add a trace, which is a graph object (bar plot) with given x and y

graph.update_layout(title = 'Pirate Scores',
                    xaxis_title = 'Pirates',
                    yaxis_title = "Score")

#graph.show() (optional line of code)

# Note that plot is interactive!!!

# Use "home" button to reset after zooming
```

```python
# Change figure type.  Can make a dictionary with plotly

graph = go.Figure() # Make our figure
graph.add_trace(go.Scatter(
                x=x1, 
                y = y1,
                mode = 'markers',
                marker = dict(size = 15, 
                              color = 'hotpink',
                              opacity = 1,
                              line = dict(width = 5, color = 'purple')) 
))
graph.update_layout(title = 'Interactive Pirate Plot', 
                    xaxis_title = 'Pirates',
                    yaxis_title = "Score",
                    width = 500,
                    height = 500)

# Use a dictionary when defining multiple characteristics of the same thing.
```

```python
# Creating a database of words from movie quotes based on a webpage of movie quotes.

df = pd.read_csv('https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv',
                    on_bad_lines = 'skip')

df
```

```python
# Make a wordcloud

text = " ".join(each for each in df.quote) # Package requires one big string of text, so
# had to separate each quote by a space so they would all be linked together.

wordcloud = WordCloud(background_color= 'white',
                      colormap = 'inferno',
                      random_state = 613).generate(text) # Setting a random seed to make sure plot looks the same each time.

# Matplotlib to put wordcloud image into figure

fig, ax = plt.subplots(figsize = (7, 3))
ax.imshow(wordcloud, interpolation = 'bilinear') # interpolation prevents dataset from looking choppy
ax.axis('off')
```

```python
# Venn Diagrams

# Define our sets (A & B)

A = set(['apple', 'banana', 'watermelon'])
B = set(['pumpkin', 'blueberry', 'apple', 'key lime'])

diagram = venn2([A, B],
                set_labels = ('Fruits', 'Pies'),
                set_colors = ('blue', 'red'),
                alpha = 0.5)
```

```python
A = set(['apple', 'banana', 'watermelon'])
B = set(['pumpkin', 'blueberry', 'apple', 'key lime'])

diagram = venn2([A, B],
                set_labels = ('Fruits', 'Pies'),
                set_colors = ('blue', 'red'),
                alpha = 0.5)

# get_label_by_id to have it list words of fruit, not just number of items.

diagram.get_label_by_id('10').set_text('\n'.join(A-B)) # left circle of diagram
diagram.get_label_by_id('11').set_text('\n'.join(A & B)) # middle of diagram
diagram.get_label_by_id('01').set_text('\n'.join(B-A)) # right

# numbers in middle are binary meaning with "1" for "on" and "0" for "off".
# So middle... both are "on".  On left, only "left" is "on".
```
